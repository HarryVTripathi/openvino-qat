```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread("../static/rz-1.jpg")
```

```python
h, w, c = img.shape
size = 12
```

### Image arithemetics

```python
add_mat = np.ones(shape=img.shape, dtype="uint8") * 100
```

```python
added = img + add_mat
plt.figure(figsize=(16, 9))
plt.imshow(cv.cvtColor(added, cv.COLOR_BGR2RGB))
plt.show()
```

#### Using `cv.add` to add images

```python
added = cv.add(img, add_mat)
plt.figure(figsize=(16, 9))
plt.imshow(cv.cvtColor(added, cv.COLOR_BGR2RGB))
plt.show()
```

This increases the brightness normally

### Bitwise operation and masking

```python
square = np.zeros((300, 300), dtype=np.uint8)
cv.rectangle(square, (50, 50), (250, 250), 255, -2)

ellipse = np.zeros((300, 300), dtype=np.uint8)
cv.ellipse(ellipse, (150, 150), (150, 150), 30, 0, 180, 255, -1)
```

```python
plt.imshow(square)
plt.show()

plt.imshow(ellipse)
plt.show()
```

```python
bit_and = cv.bitwise_and(square, ellipse)
bit_or = cv.bitwise_or(square, ellipse)
bit_xor = cv.bitwise_xor(square, ellipse)
bit_not = cv.bitwise_not(square)
```

```python
size = 20
plt.figure(figsize=(9, 12))

plt.subplot(2, 2, 1)
plt.imshow(cv.cvtColor(bit_and, cv.COLOR_BGR2RGB))

plt.subplot(2, 2, 2)
plt.imshow(cv.cvtColor(bit_or, cv.COLOR_BGR2RGB))

plt.subplot(2, 2, 3)
plt.imshow(cv.cvtColor(bit_xor, cv.COLOR_BGR2RGB))

plt.subplot(2, 2, 4)
plt.imshow(cv.cvtColor(bit_not, cv.COLOR_BGR2RGB))

plt.show()
```
